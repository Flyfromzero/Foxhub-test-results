## 测试不同的参数/batch_size下的运行时间与效果

```python
#batch_size = 4
run_time_sum = 0
batch_size = 4
for i in range(5):
    start_time = time.time()
    !python evaluate.py --checkpoint ./test_input_image/wave.ckpt --in-path ./test_input_image/test_batch_size --out-path ./test_output_image/test_batch_size --batch-size 4
    run_time = time.time() - start_time
    run_time_sum += run_time
    print("epoch %d : %s seconds"%(i,run_time))
print("--- %s seconds average if batch_size = %d on 4 jpgs for 5 epochs ---" % (run_time_sum/5,batch_size))
path = r'./test_output_image/test_batch_size'
for filename in os.listdir(path):
    img = im.imread(os.path.join(path,filename))
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
